```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
file = 'data_tp1_app.txt'
data = pd.read_csv('Archive/'+file,sep=' ', header=None, names=['label','x','y'])
```

```python
plt.figure(figsize=(20,20))
for label in np.unique(data['label']):
    plt.scatter(data[data['label']==label]['x'],data[data['label']==label]['y'])
plt.legend(["app_{}".format(l) for l in np.unique(data['label'])])
plt.show()
```
