```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('../images/cameraman.jpg',0)
plt.imshow(img,'gray')
```

```python
# Bruit S&P
def SP_noise(img,p):
    img = img.copy()
    # Matrice S&P
    SP = np.random.randint(2,size=img.shape)*255
    # Mask
    p=p/100
    mask = np.random.choice([0,1],size=img.shape,p=[1-p,p]).astype(bool)
    img[mask]=SP[mask]
    return img

plt.imshow(SP_noise(img,5),'gray')
```

```python
SP_image = SP_noise(img,5)
cv2.imwrite('../images/cameraman_s&p.jpg',SP_image)
```
