Pour la partie Implémentation, place aux filtres de Gabor. Un filtre de Gabor est un filtre linéaire dont la réponse impulsionnelle est une sinusoïde modulée par une fonction gaussienne (également appelée ondelette de Gabor). Il porte le nom du physicien anglais d'origine hongroise Dennis Gabor. De nombreux scientifiques contemporains affirment que les représentations de fréquence et d'orientation des filtres de Gabor sont similaires à celles du système visuel humain. Ils se sont avérés particulièrement appropriés pour la représentation de la texture et la discrimination. 



Les filtres de Gabor sont utilisés dans beaucoup d'applications de traitement d'image telles que le traitement d'image de document, la reconnaissance optique de caractères, d'expressions faciales et d'empreintes digitales. On raconte que la police écossaise s'est trompé pour l'arrestation de Xavier Dupont de Ligonnès car elle n'utilisait pas les filtres de Gabor dans la reconnaissance d'empreintes digitales...

## A) La construction du filtre de Gabor en détail

Un filtre de Gabor est défini par l'équation suivante :

\begin{equation*}
g(x,y;\lambda,\theta,\psi,\sigma,\gamma) = \exp\left(-\frac{x_{\theta}^2+\gamma^2y_{\theta}^2}{2\sigma^2}\right)\cos\left(2\pi\frac{x_{\theta}}{\lambda}+\psi\right)
\end{equation*}

où

\begin{equation*}x_{\theta} = x \cos\theta + y \sin\theta\,\end{equation*}

et

\begin{equation*}y_{\theta} = -x \sin\theta + y \cos\theta\,\end{equation*}

Dans cette équation, $\lambda$ représente la longueur d'onde du facteur sinusoïdal, $\theta$ représente l'orientation de la normale aux bandes parallèles d'une fonction de Gabor, $\psi$ est le décalage de phase, $\sigma$ est l'écart-type de l'enveloppe gaussienne et $\gamma$ est le rapport d'aspect spatial, et spécifie l'ellipticité du support de la fonction Gabor.

Pour un filtre de taille (X,Y), avec X et Y impairs, on considère les coordonnées (x,y) des pixels de ce filtre relativement par rapport au pixel central du kernel, qui prendra donc comme coordonnées (0,0).
